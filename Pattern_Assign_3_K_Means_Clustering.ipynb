```python
from sklearn import datasets
import numpy as np
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X.shape
```

```python
k=3
iter = 100
```

# Semi Random

```python
center = []
for i in range(0,150,int(150/k)):
  center.append(X[i])
print(center)
```

```python
for c in center:
  print(c)
```

```python
labels = np.unique(y)
print(labels)
```

```python
len(X[0])
```

```python
for b in range(iter):
  print("i am in iteration ",b)
  label=[]
  prev_center = center.copy()
  print(center)
  for sample in X:
    d = []
    min_dist = 1000000000000
    ind=-1
    for c in center:
      #if b >0:
        #print("The cluster is ",c)
      dist = sample - c
      #print("the subtract distance is ", dist)
      dist = dist**2
      #print("the power distance is ",dist)
      dist = np.sum(dist)
      #print("the sum distance is ",dist)
      dist = dist **0.5
      #print("The square distance is ",dist)
      d.append(dist)
      #print("The dist is ",dist)
    #print("The distance list is ",d)
    for i in range(len(d)):
      if d[i] < min_dist:
        min_dist = d[i]
        ind = i
        #print(ind)
    #print(d)
    m=np.min(d)
    #print(np.min(d))
    #print(d.index(m))
    label.append(ind)
  new_center = np.zeros((k,len(X[0])))
  num_samples = np.zeros(k)
  
  for i in range(len(label)):
    c = label[i]
    num_samples[c] +=1
    for j in range(len(X[0])):
      new_center[c][j]+=X[i][j]
      
  for i in range(k):
    for j in range(len(X[0])):
      if num_samples[i] == 0:
        new_center[i][j] = 0
      else:
        new_center[i][j] = new_center[i][j]/num_samples[i]
  center = new_center.copy()
  print(num_samples)
  count = 0
  for i in range(k):
    for j in range(len(X[0])):

      if prev_center[i][j] == center[i][j]:
        count += 1
  print(label)
  print(center)
  if count == k*len(X[0]) :
    break
  
```
