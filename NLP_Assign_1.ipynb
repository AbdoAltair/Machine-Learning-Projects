```python
#define NFA States
def NFA_A(array , NFA_i , que):
  if NFA_i != len(array) :
    #global NFA_Q
    print("I am in a and NFA_Q is " )
    if len(que) == 5:
      que.pop(0)
      que.append(array[NFA_i])
    else:
      que.append(array[NFA_i])
    print('i am in state a and que = ' ,que, ' and index = ' , NFA_i)
    #print(que)
    if array[NFA_i] == '0':
      print('i am in case a = 0')
      que2 = que.copy()
      NFA_A(array , NFA_i + 1 , que2)
    elif array[NFA_i] == '1':
      print('i am in case a = 1')
      que3 = que.copy()
      que4 = que.copy()
      NFA_A(array , NFA_i + 1 , que3)
      NFA_B(array , NFA_i + 1 , que4)

  

def NFA_B(array , NFA_i , que):
  if NFA_i != len(array) :
    if len(que) == 5:
      que.pop(0)
      que.append(array[NFA_i])
    else:
      que.append(array[NFA_i])
    print('i am in state b and que = ', que)
    #global NFA_Q
    #if len(NFA_Q) == 5:
    #  NFA_Q.pop(0)
    #  NFA_Q.append(array[NFA_i])
    #else:
    #  NFA_Q.append(array[NFA_i])
    #print('i am in state b and counter = ' )
    #print(NFA_Q)

    if array[NFA_i] == '0':
      que2 = que.copy()
      NFA_C(array , NFA_i +1 , que2)
    elif array[NFA_i] == '1':
      que3 = que.copy()
      NFA_B(array , NFA_i +1 , que3)
  

def NFA_C(array , NFA_i , que):
  if NFA_i != len(array) :
    if len(que) == 5:
      que.pop(0)
      que.append(array[NFA_i])
    else:
      que.append(array[NFA_i])
    print('i am in state c and que = ', que)

    
    #global NFA_Q
    #if len(NFA_Q) == 5:
    #  NFA_Q.pop(0)
    #  NFA_Q.append(array[NFA_i])
    #else:
    #  NFA_Q.append(array[NFA_i])
    #print('i am in state c and counter = '  )
    #print(NFA_Q)

    if array[NFA_i] == "0":
      que2 = que.copy()
      NFA_A(array , NFA_i +1 , que2)
    elif array[NFA_i] == "1":
      que3 = que.copy()
      que4 = que.copy()
      NFA_B(array , NFA_i +1 , que3)
      NFA_D(array , NFA_i +1 , que4)

def NFA_D(array , NFA_i , que):
  if NFA_i != len(array) :
    if len(que) == 5:
      que.pop(0)
      que.append(array[NFA_i])
    else:
      que.append(array[NFA_i])
    #global NFA_Q
    #if len(NFA_Q) == 5:
    #  NFA_Q.pop(0)
    #  NFA_Q.append(array[NFA_i])
    #else:
    #  NFA_Q.append(array[NFA_i])
    print('i am in state d and que = ' , que)
    #print(NFA_Q)

    if array[NFA_i] == "0":
      que2 = que.copy()
      que3 = que.copy()
      NFA_C(array , NFA_i +1 , que2)
      NFA_E(array , NFA_i +1 , que3)
    elif array[NFA_i] == "1":
      que4 = que.copy()
      NFA_B(array , NFA_i +1 , que4)

def NFA_E(array , NFA_i , que):
  if NFA_i != len(array) :
    if len(que) == 5:
      que.pop(0)
      que.append(array[NFA_i])
    else:
      que.append(array[NFA_i])
    #global NFA_Q
    #if len(NFA_Q) == 5:
    #  NFA_Q.pop(0)
    #  NFA_Q.append(array[NFA_i])
    #else:
    #  NFA_Q.append(array[NFA_i])
    print('i am in state e and que = ' , que)
    #print(NFA_Q)

    if array[NFA_i] == "0":
      que2 = que.copy()
      NFA_A(array , NFA_i +1 , que2)
    elif array[NFA_i] == "1":
      que3 = que.copy()
      que4 = que.copy()
      que5 = que.copy()
      NFA_B(array , NFA_i +1 , que3)
      NFA_D(array , NFA_i +1 , que4)
      NFA_F(array , NFA_i +1 , que5)

def NFA_F(array , NFA_i , que):
  global NFA_Counter
  print('f que = ' , que)
  #global NFA_Q
  if que == ["1","0","1","0","1"]:
    NFA_Counter += 1
  print('i am in state f and counter = ' , NFA_Counter)
  #print(NFA_Q)
  if NFA_i != len(array) :
    if len(que) == 5:
      que.pop(0)
      que.append(array[NFA_i])
    else:
      que.append(array[NFA_i])
  #  if len(NFA_Q) == 5:
   #   NFA_Q.pop(0)
    #  NFA_Q.append(array[NFA_i])
    #else:
     # NFA_Q.append(array[NFA_i])
    print('i am in state f and counter = ')

    if array[NFA_i] == "0":
      que2 = que.copy()
      NFA_F(array , NFA_i +1 , que2)
    elif array[NFA_i] == "1":
      que3 = que.copy()
      NFA_F(array , NFA_i +1 , que3)

```

```python
def DFA_A(array , DFA_i):
  if DFA_i != len(array) :
    global DFA_Q
    if len(DFA_Q) ==5:
      DFA_Q.pop(0)
      DFA_Q.append(array[DFA_i])
    else:
      DFA_Q.append(array[DFA_i])

    if array[DFA_i] == '0':
      DFA_A(array , DFA_i +1)
    elif array[DFA_i] == '1':
      DFA_AB(array , DFA_i + 1)

def DFA_AB(array , DFA_i):
  if DFA_i != len(array) :
    global DFA_Q
    if len(DFA_Q) ==5:
      DFA_Q.pop(0)
      DFA_Q.append(array[DFA_i])
    else:
      DFA_Q.append(array[DFA_i])
    if array[DFA_i] == '0':
      DFA_AC(array , DFA_i +1)
    elif array[DFA_i] == '1':
      DFA_AB(array , DFA_i +1)
      

def DFA_AC(array , DFA_i):
  if DFA_i != len(array) :
    global DFA_Q
    if len(DFA_Q) ==5:
      DFA_Q.pop(0)
      DFA_Q.append(array[DFA_i])
    else:
      DFA_Q.append(array[DFA_i])
    if array[DFA_i] == '0':
      DFA_A(array , DFA_i +1)
    elif array[DFA_i] == '1':
      DFA_ABD(array , DFA_i +1)

def DFA_ABD(array , DFA_i):
  if DFA_i != len(array) :
    global DFA_Q
    if len(DFA_Q) ==5:
      DFA_Q.pop(0)
      DFA_Q.append(array[DFA_i])
    else:
      DFA_Q.append(array[DFA_i])
    if array[DFA_i] == '0':
      DFA_ACE(array ,DFA_i +1)
    elif array[DFA_i] == '1':
      DFA_AB(array , DFA_i +1)

def DFA_ACE(array , DFA_i):
  if DFA_i != len(array) :
    global DFA_Q
    if len(DFA_Q) ==5:
      DFA_Q.pop(0)
      DFA_Q.append(array[DFA_i])
    else:
      DFA_Q.append(array[DFA_i])
    if array[DFA_i] == '0':
      DFA_A(array , DFA_i +1)
    elif array[DFA_i] == '1':
      DFA_ABDF(array , DFA_i +1)

def DFA_ABDF(array , DFA_i):
  global DFA_Counter
  global DFA_Q
  
  if DFA_Q == ["1","0",'1','0','1']:
    DFA_Counter += 1
  if DFA_i != len(array) :
    if len(DFA_Q) ==5:
      DFA_Q.pop(0)
      DFA_Q.append(array[DFA_i])
    else:
      DFA_Q.append(array[DFA_i])
    if array[DFA_i] == '0':
      DFA_ACEF(array , DFA_i +1)
    elif array[DFA_i] == '1':
      DFA_ABF(array , DFA_i +1)

def DFA_ACEF(array , DFA_i):
  if DFA_i != len(array) :
    global DFA_Q
    if len(DFA_Q) ==5:
      DFA_Q.pop(0)
      DFA_Q.append(array[DFA_i])
    else:
      DFA_Q.append(array[DFA_i])
    if array[DFA_i] == '0':
      DFA_AF(array , DFA_i +1)
    elif array[DFA_i] == '1':
      DFA_ABDF(array , DFA_i +1)

def DFA_ABF(array , DFA_i):
  if DFA_i != len(array) :
    global DFA_Q
    if len(DFA_Q) ==5:
      DFA_Q.pop(0)
      DFA_Q.append(array[DFA_i])
    else:
      DFA_Q.append(array[DFA_i])
    if array[DFA_i] == '0':
      DFA_ACF(array , DFA_i +1)
    elif array[DFA_i] == '1':
      DFA_ABF(array , DFA_i +1)

def DFA_AF(array , DFA_i):
  if DFA_i != len(array) :
    global DFA_Q
    if len(DFA_Q) ==5:
      DFA_Q.pop(0)
      DFA_Q.append(array[DFA_i])
    else:
      DFA_Q.append(array[DFA_i])
    if array[DFA_i] == '0':
      DFA_AF(array , DFA_i +1)
    elif array[DFA_i] == '1':
      DFA_ABF(array , DFA_i +1)

def DFA_ACF(array , DFA_i):
  if DFA_i != len(array) :
    global DFA_Q
    if len(DFA_Q) ==5:
      DFA_Q.pop(0)
      DFA_Q.append(array[DFA_i])
    else:
      DFA_Q.append(array[DFA_i])
    if array[DFA_i] == '0':
      DFA_AF(array , DFA_i +1)
    elif array[DFA_i] == '1':
      DFA_ABDF(array , DFA_i +1)
```
