```python
#import cv2, numpy and matplotlib libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread("/content/drive/MyDrive/circuit.jpg")
#Displaying image using plt.imshow() method
plt.imshow(img)
 
#hold the window
```

```python
laplace4 = np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
mask4 = cv2.filter2D(img, -1, laplace4, borderType=cv2.BORDER_CONSTANT) 
plt.imshow(mask4)
```

```python
imglaplace4 = img+mask4
plt.imshow(imglaplace4)
```

```python
laplace8 = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
mask8 = cv2.filter2D(img,-1,laplace8,borderType=cv2.BORDER_CONSTANT)
plt.imshow(mask8)
```

```python
imglaplace8 = img+mask8
plt.imshow(imglaplace8)
```
