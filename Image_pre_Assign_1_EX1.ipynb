```python
gray_image = cv2.imread('/content/drive/MyDrive/lake-192990__340.jpg',0)
```

```python
plt.imshow(gray_image)
```

```python
histg = cv2.calcHist([gray_image],[0],None,[256],[0,256]) 
```

```python
plt.plot(histg)
plt.show()
```

```python

data_shift = np.where((255-gray_image) < 50 , 255 , gray_image+50)

```

```python
histg_shifted = cv2.calcHist([data_shift],[0],None,[255],[0,255]) 
```

```python
plt.imshow(data_shift)
```

```python
plt.plot(histg_shifted)
plt.show()
```
